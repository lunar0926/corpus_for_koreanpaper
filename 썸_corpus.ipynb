# 챗봇 문답 페어

```python
# 챗봇 문답 페어 
chatbot_corpus = Korpora.load("korean_chatbot_data")
```

```python
chatbot_text = chatbot_corpus.get_all_texts()
```

```python
len(chatbot_text)
```

```python
# chatbot_text 표현 체크
chatbot_text_some = []
for text in chatbot_text:
  if '썸' in text:
    chatbot_text_some.append(text)
  
```

```python
chatbot_text_some
```

```python
# '썸'이 포함된 질문 텍스트의 개수 
len(chatbot_text_some)
```

```python
chatbot_pair = chatbot_corpus.get_all_pairs()
```

```python
# chatbot_pair 표현 체크
chatbot_pair_some = []
for text in chatbot_pair:
  if '썸' in text:
    chatbot_pair_some.append(text)
```

```python
chatbot_pair_some
```

```python
len(chatbot_pair_some)
```
